# Story Weaver

## Objective
To create a cohesive and imaginative story using words on the cards.

## Players
2 to 6 players (optimal for engaging storytelling).

## Game Components
1. **First Cards (20 cards):** Each card features a unique word and an associated image. These cards serve as the story's starting point.
2. **Middle Cards (100 cards):** Each card has a word and an image. These cards build the body of the story.
3. **Final Cards (20 cards):** Similar to the First and Middle cards, but these are used to conclude the story.

## Setup
1. Each player should open page of this Google Collaboratory document.
2. Each player is dealt 7 cards: 5 Middle cards and 2 Final cards.



## Gameplay
1. **Starting the Game:** The youngest player begins by drawing a First card, revealing the word and picture.
2. **Storytelling:** The starting player creates the first sentence of the story, incorporating the word from the First card.
3. **Continuing the Story:** The same player or the next player (depending on the chosen rule variation) adds to the story by playing a Middle card, using its word in a new sentence. The story should flow logically and creatively.
4. **Ending the Story:** Players use their Final cards to conclude the story. These cards can only be played once a player has used all their Middle cards.
5. **Passing the Turn:** If a player cannot continue the story or the story becomes illogical or repetitive, they pass their turn, and the next player continues.

## Winning
The first player to use all their Middle cards and one Final card to end the story wins.

## Rule Variations
1. **Turn-Based:** Players take turns adding to the story.
2. **Free-Form:** Players can interrupt to add to the story, promoting faster gameplay and spontaneity.

```python
import os
from IPython.display import Image, display
import random
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

!git clone https://github.com/matthewthefool/In-the-beginning-was-the-Card.git

def Initiate_game():

  global Firstcards, Middlecards, Finalcards, Firstcards_path, Middlecards_path, Finalcards_path
  # Define the directory path where each card type is
  Middlecards_path = "/content/In-the-beginning-was-the-Card/Cards/Middlecards"
  Finalcards_path = "/content/In-the-beginning-was-the-Card/Cards/Finalcards"
  Firstcards_path = "/content/In-the-beginning-was-the-Card/Cards/Firstcards"

  # Deal player 5 random Middlecards
  all_files = os.listdir(Middlecards_path)
  Middlecards = random.sample(all_files, 5)

  # Deal player 2 random Finalcards
  all_files = os.listdir(Finalcards_path)
  Finalcards = random.sample(all_files, 2)

  # Lay 1 Firstcard on the table
  all_files = os.listdir(Firstcards_path)
  Firstcard = random.sample(all_files, 1)

  print(f"Here the game begins!:")
  print(Firstcard[0])
  display(Image(filename=os.path.join(Firstcards_path, Firstcard[0])))


def Display_Cards():
    if 'Middlecards' not in globals():
        print("You have to initiate the game first by running code tile with 'Initiate_game()' written in it.")
        return

    print(f"Middlecards:")
    image_paths = [Middlecards_path + '/' + Middlecard for Middlecard in Middlecards]
    Middlecards_names = [card.split('.')[0] for card in Middlecards]

    # Initialize an empty list to store the resized images
    resized_images = []

    for image_path in image_paths:
        img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

        # Resize the image to the desired width and height
        img = cv2.resize(img, (150, 150))

        # Append the resized image to the list
        resized_images.append(img)

    # Concatenate the resized images horizontally
    horizontal_concatenation = np.hstack(resized_images)

    # Define font properties for the text
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.6  # Adjust the font size (smaller)
    font_thickness = 2  # Adjust the font thickness
    text_color_white = (255, 255, 255)  # White color
    text_color_black = (0, 0, 0)  # Black color

    # Add the black outline to the text
    for i, image_name in enumerate(Middlecards_names):
        x = i * 150  # X-coordinate for the text
        y = 30  # Y-coordinate for the text (adjust as needed)

        # Draw the black outline
        cv2.putText(horizontal_concatenation, image_name, (x, y), font, font_scale, text_color_black, 4, cv2.LINE_AA)

        # Draw the white text on top of the black outline
        cv2.putText(horizontal_concatenation, image_name, (x, y), font, font_scale, text_color_white, font_thickness, cv2.LINE_AA)

    # Display the horizontally concatenated image with text
    cv2_imshow(horizontal_concatenation)

    print(f"Finalcards:")
    image_paths = [Finalcards_path + '/' + Finalcard for Finalcard in Finalcards]
    Finalcards_names = [card.split('.')[0] for card in Finalcards]
    # Initialize an empty list to store the resized images
    resized_images = []

    for image_path in image_paths:
        img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

        # Resize the image to the desired width and height
        img = cv2.resize(img, (150, 150))

        # Append the resized image to the list
        resized_images.append(img)

    # Concatenate the resized images horizontally
    horizontal_concatenation = np.hstack(resized_images)

    # Define font properties for the text
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.6  # Adjust the font size (smaller)
    font_thickness = 2  # Adjust the font thickness
    text_color_white = (255, 255, 255)  # White color
    text_color_black = (0, 0, 0)  # Black color

    # Add the black outline to the text
    for i, image_name in enumerate(Finalcards_names):
        x = i * 150  # X-coordinate for the text
        y = 30  # Y-coordinate for the text (adjust as needed)

        # Draw the black outline
        cv2.putText(horizontal_concatenation, image_name, (x, y), font, font_scale, text_color_black, 4, cv2.LINE_AA)

        # Draw the white text on top of the black outline
        cv2.putText(horizontal_concatenation, image_name, (x, y), font, font_scale, text_color_white, font_thickness, cv2.LINE_AA)

    # Display the horizontally concatenated image with text
    cv2_imshow(horizontal_concatenation)


def Used_card():
  if 'Middlecards' not in globals():
        print("You have to initiate the game first by running code tile with 'Initiate_game()' written in it.")
        return
  # Ask the user for input
  card_to_delete = input("Which card have you used? ")

  # Add ".png" to the card name provided by the user
  card_to_delete_with_extension = card_to_delete + ".jpg"

  # Check if the card with ".png" extension is in the Middlecards list
  if card_to_delete_with_extension in Middlecards:
      # Remove the card from the list
      Middlecards.remove(card_to_delete_with_extension)
      print(f"{card_to_delete} has been removed from the Middlecards list.")
  else:
      print(f"{card_to_delete} is not in the Middlecards list.")

def Pass():
  if 'Middlecards' not in globals():
        print("You have to initiate the game first by running code tile with 'Initiate_game()' written in it.")
        return

  global Middlecards
  # Define the directory path
  directory_path = "/content/drive/MyDrive/StoryWeaver/Cards/Middlecards"

  # Use os.listdir to get a list of all files and directories in the specified directory
  all_files_and_directories = os.listdir(directory_path)

  # Filter the list to include only files (exclude directories)
  all_files = [f for f in all_files_and_directories if os.path.isfile(os.path.join(directory_path, f))]

  # Use random.sample to select 5 random entries from the list without duplicates
  Middlecards = Middlecards + random.sample(all_files, 1)

def Victory():
  if 'Middlecards' not in globals():
        print("Your really thought of winning before even starting the game?")
        return
  # Ask the user for input
  Winner = input("What is the name of the lucky one? What is the name of the winner? ")
  print(f"You, {Winner}, are the real skazocznik, better then Hanz Kristien Alexandrien.")
```
